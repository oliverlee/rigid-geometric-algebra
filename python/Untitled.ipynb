```python
import numpy as np

import pyvista as pv
pv.set_jupyter_backend('trame') 

rng = np.random.default_rng(seed=0)
points = rng.random((1000, 3))

rgba = points - points.min(axis=0)
rgba /= rgba.max(axis=0)
pv.plot(
    points,
    scalars=rgba,
    render_points_as_spheres=True,
    point_size=20,
    rgba=True,
)
```
